# Introduction

Sentiment Analysis is a Natural Language Processing (NLP) technique used to determine whether a piece of text expresses a positive, negative, or neutral sentiment.

This project analyzes customer reviews from Amazon products to understand customer opinions, identify sentiment patterns, and generate valuable business insights.

# Problem Statement

Customer reviews contain valuable information about product quality and customer satisfaction. Manually analyzing thousands of reviews is difficult and time-consuming.

This project aims to automatically classify customer reviews into positive, negative, and neutral sentiments to understand customer opinions and support business decision-making.

# Objectives

• Analyze customer reviews using sentiment analysis.

• Classify reviews as Positive, Negative, or Neutral.

• Apply Natural Language Processing techniques.

• Understand customer opinions and trends.

• Generate insights for business and marketing decisions.

# Dataset Description

The dataset contains Amazon customer reviews with the following features:

• ReviewID

• Category

• Review

• Rating

• Sentiment

# Import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from textblob import TextBlob
from wordcloud import WordCloud

import warnings
warnings.filterwarnings('ignore')
```

# Load Dataset

```python
df = pd.read_csv("Amazon_Reviews_Dataset.csv")

df.head()
```

# Data Understanding

```python
df.shape
df.info()
df.describe()
```

# Data Quality Check

```python
df.isnull().sum()
```

# Duplicate Values

```python
df.duplicated().sum()
```

# Sentiment Distribution

```python
plt.figure(figsize=(6,4))
sns.countplot(x='Sentiment', data=df)
plt.title("Sentiment Distribution")
plt.show()
```

# Ratings Distribution

```python
plt.figure(figsize=(6,4))
sns.countplot(x='Rating', data=df)
plt.title("Ratings Distribution")
plt.show()
```

# Category-wise Reviews

```python
plt.figure(figsize=(8,5))
sns.countplot(x='Category', data=df)
plt.title("Reviews by Category")
plt.xticks(rotation=20)
plt.show()
```

# Sentiment by Category

```python
plt.figure(figsize=(8,5))
sns.countplot(x='Category',
              hue='Sentiment',
              data=df)

plt.xticks(rotation=20)
plt.title("Category-wise Sentiment")
plt.show()
```

# Word Cloud

```python
text = " ".join(df["Review"])

wordcloud = WordCloud(
    width=800,
    height=400,
    background_color="white"
).generate(text)

plt.figure(figsize=(10,5))
plt.imshow(wordcloud)
plt.axis("off")
plt.title("Word Cloud")
plt.show()
```

# NLP Sentiment Validation

```python
def get_polarity(text):
    return TextBlob(text).sentiment.polarity

df["Polarity"] = df["Review"].apply(get_polarity)

df.head()
```
